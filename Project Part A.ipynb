```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


filepath = "train.csv"
housingdata = pd.read_csv(filepath,sep=',');

fig = plt.figure()

ax = fig.add_subplot(111,projection='3d')

ax.scatter(housingdata['YearRemodAdd'],housingdata['OverallQual'],housingdata['TotRmsAbvGrd'])
ax.set_title('The Relation between Year Remodeled, Overall Quality and Number of Total Rooms')
ax.set_xlabel('Year Remodeled')
ax.set_ylabel('Overall Quality')
ax.set_zlabel('Total Number of Rooms')
plt.show(fig)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
filepath = "train.csv"
housingdata = pd.read_csv(filepath,sep=',');
subset_data = housingdata[['Neighborhood', 'SalePrice']]
grouped_data = []
titles = subset_data['Neighborhood'].unique()
for l in titles:
     grouped_data.append((subset_data[subset_data['Neighborhood'] == l])['SalePrice'].values)

fig2 = plt.figure(figsize=(80,60))

ax2 = fig2.add_subplot(111)
ax2.boxplot(grouped_data, labels=titles)
ax2.tick_params(axis='both', which='major', labelsize=40)
plt.title('Sale Prices of Houses Across Several Neighborhoods in Ames, Iowa', fontsize=100)
plt.xticks(rotation=-70)
plt.show(fig2)
```
